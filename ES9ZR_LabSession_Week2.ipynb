**SciPy**

SciPy is a fundamental Python library designed for scientific computing. It builds upon the core functionality of NumPy, offering modules for optimization, integration, interpolation, eigenvalue problems, algebraic equations, and more. Widely used in academic research and industry, it provides tools for tasks across many branches of science and engineering. With its robust collection of algorithms and data structures, SciPy enhances Python's capabilities for mathematical and numerical computations. Integral to the Python scientific stack, it is commonly used alongside libraries like NumPy, Matplotlib, and Pandas, facilitating a comprehensive environment for data analysis, visualization, and algorithmic exploration.

image.png

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

data = np.genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', delimiter=',', skip_header=0)  #
####delet the 5th nan column
data = data[:,0:4]

desc_stats = stats.describe(data)

print(desc_stats)


print('##################################################Seperation###########################################')

# define data
x = data[:,0]  #### x=
ydata = data[:,2]
def func(x, a):
    return a * x


# use curve_fit function to
params, params_covariance = curve_fit(func, x, ydata)

# visulization
plt.figure()
plt.scatter(x, ydata, label='IrisData')
plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.plot(x, func(x, params[0]), label='Fitted function')
plt.legend(loc='best')
plt.show()
```

**Matplotlib**

Matplotlib is a comprehensive Python library for creating static, interactive, and animated visualizations. Established by John D. Hunter in 2003, it has since become one of the most widely used plotting libraries in the Python ecosystem. Built on NumPy arrays, Matplotlib integrates seamlessly with Python's scientific computing stack, providing a flexible platform for producing a myriad of plots and figures, from simple line graphs to intricate 3D visualizations. Its intuitive API allows beginners to generate basic plots quickly, while its extensive customization options cater to advanced users' needs. Often utilized alongside data manipulation libraries like Pandas, Matplotlib is foundational for data visualization in Python.

image.png

```python
## Load libraries

import numpy as np
import matplotlib.pyplot as plt

## load Iris dataset
data = np.genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', delimiter=',', skip_header=0)  #

sepal_length = data[:, 0]
sepal_width = data[:, 1]
petal_length  = data[:, 2]




# scatter plot
plt.figure(figsize=(5, 5))
plt.scatter(range(len(sepal_length)), sepal_length)
plt.title("Scatter plot of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()

# boxplot
plt.figure(figsize=(5, 5))
datainBox = [sepal_length,sepal_width]
plt.boxplot(datainBox)
plt.title("Box plot of Sepal Length/Width")
plt.ylabel("Sepal Length/Width (cm)")
plt.show()






```

```python

###### histgram

# plot
plt.figure(figsize=(5, 5))
fig, ax = plt.subplots()
ax.hist(sepal_length, bins=8, linewidth=0.5, edgecolor="white")
plt.title("Histgram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Counts")




##### 3D graph
plt.figure(figsize=(20, 20))
plt.style.use('_mpl-gallery')
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(sepal_length, sepal_width, petal_length)

ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length')
plt.show()
```
