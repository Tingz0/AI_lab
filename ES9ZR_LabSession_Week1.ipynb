We often print a combination of pre-defined strings and some variables, and variables can be formatted and inserted into such strings using the `%` operator and various formatting commands like `%s` for strings, `%f` for floating point numbers, etc.  An alternative way of achieving the same results, which some users prefer, is the use of "f-strings". Practicing the **"print"** command is a critical step in your Python learning as it would help you check the variables, especially in debugging.

```python
from datetime import date
t = date.today()
print("Hello World. Today is %s." % t)
x = 0.123456
print("x is %s." % x)
print("x is %.3f to 3 decimal places." % x)
### And here is another way to print the same strings
print(f"Hello World. Today is {t}.")
print(f"x is {x}.")
print(f"x is {x:.3} to 3 decimal places.")
```

## Basic data types

Some of the **basic data types** in python include **integers**, **Booleans** **floats** and **strings**.  These can be grouped together into **lists** (and other list-like objects such as **tuples**, **dictionaries**, and **arrays** -- but more on those later).

```python
#Let's start from a simple integer addition problem
x = 2023
y = 1
print(x + y)
```

Whereas a single equals sign (`=`) is used to assign a value to a variable, a double equals sign (`==`) is used to test for equality.  That is, for two variables `a` and `b`, the expression `a == b` is a Boolean variable, either `False` ot `True`.  For an even more strict test of equality, use the keyword `is`.

```python
a=1
b=1.000
print(f"a == b? {a == b}")  ### We use f-string here, which could alow us evaluate the command inside the {}.
print(f"a is b? {a is b}")
```

The difference between `a == b` and `a is b` above is down to the fact that Python treats integers and floats as being different even if they have the same numerical value, because they are different data types.  Another good example of the difference between `==` and `is` comes up when we compare the integers `0` and `1` to the Boolean values `False` and `True`:

```python
print(f"0 == False? {0 == False}")
print(f"1 == True? {1 == True}")
print(f"0 is False? {0 is False}")
print(f"1 is True? {1 is True}")
print(f"2 == True? {2 == True}") ## NB Python differs from many other languages here!
```

Booleans can be combined using fairly obvious logical connectives like `and`, `or`, `not` etc. This characteristic could help with your "if" and "while" logic command.

Strings consist of zero or more characters enclosed by matching single or double quotation marks.  Whether you use single or double quotes is a matter of choice, but they do need to match;  to use the outermost style of quotes inside the string, you will need to escape the inner quotes with a backslash, as in the string `s2` below.

```python
s1 = "The 'quick' brown fox"
s2 = "jumped over the \"lazy\" dog."
s = s1 + " " + s2 ## Let's concatenate the two strings with a space in between them
print(s) ## and print the result
s = " ".join([s1, s2,]) ## Let's concatenate the two strings in a better way
print(s) ## and print the result
```

Strings can be "sliced" to yield substrings such as the first character, the first `n` characters, the last character, every other character, etc.  The general format is `string[first:last:stride]`, with all three arguments being optional, as the following examples show:

```python

print(s[0]) ## The first (zeroth) character of the string s. Python start from '0' rather than '1'
print(s[0:10]) ## The first 10 characters of the string s
print(s[-1]) ## The last character of the string s
print(s[::2]) ## Every other character in s
print(s[1:8:2]) ## Every other character in s, starting with the first, up to the eighth
```

One nice feature of strings in Python is that checking for whether or not one string is a substring of another can be done using the user-friendly keyword `in`:

```python
print(f"'The' in s? {'The' in s}")
print(f"'Then' in s? {'Then' in s}")
```

## Basic plotting in matplotlib

There are many data visualisation tools available for Python, but the workhorse and best place to start is `matplotlib`.  The `matplotlib` module offers a various visualisation tools for plotting curves, surfaces, and the like.  There are other libraries that are better suited to more exotic data types;  for example, the `geopandas` module is is well suited to plotting geographical data on maps.

```python
import matplotlib.pyplot as plt

X = range(0,100,1)
Y = range(0,200,2)
Z =  [X * X for X in X]


#### This is a line chart
# print('This is a line chart')
fig, ax = plt.subplots()
ax.plot(X, Y)
ax.set_xlabel("x")
ax.set_ylabel("y=2x");

###This is a box chart
# print('This is a box chart')
fig, ax = plt.subplots()
plt.boxplot(Z)

###This is a heat map
# print('This is a heat map')
fig, ax = plt.subplots()
plt.imshow([X,Y], cmap='rainbow', interpolation='bicubic')

# print('This is a bar chart')
fig, ax = plt.subplots()
plt.bar(Y, X)
```
